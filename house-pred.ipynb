```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

dataset='/kaggle/input/home-data-for-ml-course/train.csv'
data=pd.read_csv(dataset)
data.head()
```

```python
features=['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X=data[features]
Y=data.SalePrice
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y ,random_state=1)
```

```python
#model
model=RandomForestRegressor(random_state=1)
model.fit(X_train,Y_train)
predictions=model.predict(X_test)
mae_comp=mean_absolute_error(predictions,Y_test)
```

```python
model_full=RandomForestRegressor(random_state=1)
model_full.fit(X,Y)
```

```python
test_data='/kaggle/input/home-data-for-ml-course/test.csv'
tests_data=pd.read_csv(test_data)
X_tests=tests_data[features]
pred=model_full.predict(X_tests)
```

```python
output=pd.DataFrame({'ID':tests_data.Id,'saleprice':pred})
output.to_csv('submission.csv',index=False)
```
